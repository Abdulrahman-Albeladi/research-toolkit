Pangram classified about <strong>63.7%</strong> of the analyzed JavaScript/Web Development code content as AI-like, and more than half of submissions were AI-dominant. </p>

The detector did not meaningfully identify an AI-assisted middle category: the AI-assisted fraction was essentially zero across submissions.
</p>

<p>
These results indicate that AI-like characteristics were widespread in the coding submissions and were more prevalent in the Second Semester sample. </p>

## 1. Detector Calibration on the Testing Sample

This section evaluates detector performance on a **controlled testing sample** where the expected label is already known.

### Goal
The goal is to estimate how accurate each detector is on:
- **AI-Free** texts, which should be classified as **human**,
- **Humanized** texts, which should still be treated as **AI-derived**, and
- **AI-Full** texts, which should be classified as **AI**.

### Why this matters
These calibration results provide the justification for later weighting the detectors differently. A detector that performs better on the testing sample is given more influence in the combined AI-rate used for the full assignment dataset.

### What the next code block does
The next block:
- mounts Google Drive,
- loads the testing-sample workbook,
- standardizes detector labels,
- computes per-tool and per-set accuracy,
- computes average AI-rate values, and
- saves a clean TXT summary report for later reference.

### 1.1 Interpretation of the Testing-Sample Results

The testing sample shows that all four detectors were strong at identifying the clearly human-written **AI-Free** files and the clearly AI-generated **AI-Full** files. Performance dropped much more sharply on the **Humanized** files, which were the hardest category to classify correctly. This means post-editing or humanization made AI-origin text substantially harder for the detectors to identify.

Across all three reference sets, **GPTZero** performed best overall, followed by **Sapling**, then **Pangram**, while **Isgen** showed the weakest performance. The largest weakness for all tools appeared in the Humanized texts, where accuracy fell the most. In practical terms, this supports the decision to treat detector output cautiously when the text may have been revised or humanized after AI generation, and it also supports using **weighted detector contributions** instead of a simple equal average in later analyses.

- For coding assignments, the combined rate uses **Pangram only**, as Pangram is the only detector available in that workflow branch.

### 2.1 Interpretation of the Assignment-Level Summary

The full assignments dataset was analyzed using a **combined AI-rate** based on weighted detector contributions. The weights came from the testing-sample calibration results, so detectors with stronger measured performance contributed more to the final combined score. For coding assignments, the combined score relied on Pangram alone because the other writing-focused detectors were not available for that task type.

At the dataset level, the combined AI-rate showed substantial variation across files, which means the corpus was not homogeneous in its detector profile. The section also produced descriptive comparisons by university, language, major, year, and assignment type, along with relationships between AI-rates and word count. In addition, it identified files with strong disagreement between detectors and exported both a conflict table and matching raw PDFs for manual review. ## 3. Detailed Conflict Analysis Between Detectors
